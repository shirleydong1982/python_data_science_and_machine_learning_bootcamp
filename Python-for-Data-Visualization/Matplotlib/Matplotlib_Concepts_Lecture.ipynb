# Object oriented plotting method: create a plotting method

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
# two figures in one axes

fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
#axes2=fig.add_axes([0.5,0.15,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
# 
```

```python
# subplots

fig,axes=plt.subplots(nrows=1,ncols=2)  

"""
# axes can be iterated
for current_ax in axes:
    current_ax.plot(x,y)
"""


axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
    
plt.tight_layout()  # solve the overlapping issue
# fig.tight_layout()
```

# Figure Size and DPI

```python
# use parameter figsize=() to specify the figure size
fig=plt.figure(figsize=(8,2),dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig=plt.figure(0)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
# legend

fig=plt.figure(0)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Sqaured')
ax.plot(x,x**3,label='X Cubed')
#ax.legend(loc=0)  # need to specify the label in ax.plot, in order for ax.legend() to work
ax.legend(loc=(0.1,0.1))  # can use tuple in loc to specify the lower left corner coordinate
```

# Plot Appearance

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
#ax.plot(x,y,color='orange')
ax.plot(x,y,color='#FF8C00')  # RGB Hex code
```

```python
# linewidth and linestyle

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

# ax.plot(x,y,color='purple',linewidth=3,alpha=0.5)    # alpha is transparency
# ax.plot(x,y,color='purple',lw=3,alpha=0.5)  
ax.plot(x,y,color='purple',lw=3,alpha=0.5,linestyle='steps')   #-,--,-.,:,steps
ax.plot(x,y,color='purple',lw=3,alpha=0.5,ls='steps')   #-,--,-.,:,steps
```

```python
# markers

# linewidth and linestyle

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

# ax.plot(x,y,color='purple',linewidth=3,alpha=0.5)    # alpha is transparency
# ax.plot(x,y,color='purple',lw=3,alpha=0.5)  
ax.plot(x,y,color='purple',lw=0.5,alpha=0.5,ls='-',marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')  

```

```python
# xlimit and ylimit

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2,alpha=0.5,ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
