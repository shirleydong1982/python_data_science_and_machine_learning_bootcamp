# Pandas Built-in Data Visualization

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df1=pd.read_csv('df1',index_col=0)
```

```python
df1.head()
```

```python
df2=pd.read_csv('df2')
```

```python
df2.head()
```

```python
df1['A'].hist(bins=30)
```

```python
df1['A'].plot(kind='hist',bins=30) # use kind to specify what plot you want
```

```python
df1['A'].plot.hist(bins=30)  # this is the method we prefer to use when we use pandas for the plotting
```

```python
df2.head()
```

```python
df2.plot.area(alpha=0.4)  #alpha for transparancy
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

```python
df1['A'].plot.hist(bins=50)
```

```python
df1.head()
```

```python
df1.plot.line(x=df1.index,y='B',figsize=(12,3),lw=1)
```

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*100,figsize=(10,8))
```

```python
df2.plot.box()
```

```python
df=pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
```

```python
df.head()
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,figsize=(10,8),cmap='coolwarm')
```
