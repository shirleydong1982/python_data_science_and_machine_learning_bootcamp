# Categorical Plots

```python
"""
Mainly for categorical variables
sns.barplot: a little different from barplot in R, you have to specify y (a continuous variable), x is the categorical variable
sns.countplot: is the real bar plot, y is the count of the x
sns.boxplot: similar to barplot above, x is categorical, y is continuous
sns.violinplot: similar to historgram, used for a single column in the data frame
sns.stripplot: jittered plot
sns.swarmplot: recommend not to use this

sns.factorplot(kind='box/violin/bar/count/strip/swarm'): can use kind argument to generate all the plots above
"""
```

```python
import seaborn as sns
%matplotlib inline

tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
# boxplot()
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
# violin plot()
sns.violinplot(x='day',y='total_bill',data=tips)  # provide boxplot, plus barplot
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)  # scatter plot based on categorical variable
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)  # scatter plot based on categorical variable, jitter style
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')  # scatter plot based on categorical variable, jitter style
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)  # scatter plot based on categorical variable, jitter style
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)  #combine stripplot, violinplot. however, drawback: did not show the scale well
```

```python
sns.violinplot(x='day',y='total_bill',data=tips) #combine violinplot with swarmplot
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')  
```

```python
# factorplot: can be used to generate all the plots above, by changing the argument kind
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='box')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```
