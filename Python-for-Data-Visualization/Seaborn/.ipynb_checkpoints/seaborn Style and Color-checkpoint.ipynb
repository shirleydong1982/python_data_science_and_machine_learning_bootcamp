# Seaborn Style and Color

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

# set style of the plot

```python
sns.set_style('darkgrid')  # white, ticks,darkgrid,whitegrid
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')  # white, ticks,darkgrid,whitegrid
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True)
```

# figure size: using figsize

```python
plt.figure(figsize=(12,3))  # overwrite the fig size in seaborn
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('poster')  # change the set_context to poster, it will change the size of the figure and the font of the label
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('poster',font_scale=1.5)  
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('notebook')  
sns.countplot(x='sex',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')  # magma
```

```python
# https://matplotlib.org/examples/color/colormaps_reference.html for the color in palette
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='seismic') 
```
