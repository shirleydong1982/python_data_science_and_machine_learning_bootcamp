# Regression Plots

```python
# lmplot(): a scatter plot with a regression line on top of it. can use hue, col,row to have facetgrid function
```

```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips) # a scatter plot with a regression line on top of it
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex') # use hue on categorical variable
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100}) # use hue on categorical variable
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')  # this is facetgrid, instead of using hue
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex')  
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex')    # aspect ratio is wrong
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,size=8)
```
