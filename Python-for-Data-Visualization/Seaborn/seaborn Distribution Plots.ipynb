# Distribution Plots

```python
"""
Mainly for continuous variables
sns.distplot: histogram + kde plot for a single column in the data frame
sns.jointplot: scatter plot/regression plot + histogram for two columns in the data frame
sns.pairplot: plot every single combination of columns in the data frame 
sns.rugplot: similar to historgram, used for a single column in the data frame
sns.kdeplot:
"""
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)  # histogram + kde
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)  # for two variables, scatter plot + distribution plot
```

```python
 # for two variables, distribution plot + hex (use kind argument to change what happen in the middle of the plot)
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex') 
```

```python
 # for two variables, distribution plot + regression line (use kind argument to change what happen in the middle of the plot)
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') 
```

```python
 # for two variables, distribution plot + kde (use kind argument to change what happen in the middle of the plot)
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde') 
```

```python
# pairplot: plot for every single combination of variables in your dataframe
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')  # hue will take a categorical variable
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')  # hue will take a categorical variable. same as aes(fill=variable) in ggplot2
```

```python
sns.rugplot(tips['total_bill'])  # same as distplot, pass a single column in the data frame
```

```python
sns.distplot(tips['total_bill'])  # kernal density estimation
```

```python
sns.kdeplot(tips['total_bill'])
```
